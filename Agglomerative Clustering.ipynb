# Agglomerative Clustering

## **What is Agglomerative Clustering?**

Agglomerative clustering is an **unsupervised machine learning technique** used to group similar data points into clusters. It follows a **bottom-up approach**, where each data point starts in its own cluster, and the closest clusters are merged step-by-step. The result of this clustering method is a **dendrogram**, a tree-like diagram that illustrates how the data points are merged into clusters. This technique is particularly useful for uncovering a **hierarchical structure** in the data and gaining a deeper understanding of its underlying patterns.

## **Key Components of Agglomerative Clustering**



### 1. Distance or Similarity Metric

The distance or similarity metric quantifies how similar or dissimilar two data points or clusters are. Common metrics include:

- **Euclidean Distance**: For two points $x$ and $y$ in an $n$-dimensional space, the distance $d(x, y)$ is defined as:
  $$
  d(x, y) = \sqrt{\sum_{i=1}^{n} (x_i - y_i)^2}
  $$

- **Manhattan Distance**: For the same points $x$ and $y$, the distance $d(x, y)$ is defined as:
  $$
  d(x, y) = \sum_{i=1}^{n} |x_i - y_i|
  $$

- **Cosine Similarity**: For two vectors $x$ and $y$, the similarity is defined as:
  $$
  \text{similarity}(x, y) = \frac{x \cdot y}{\| x \| \| y \|}
  $$
  where $x \cdot y$ is the dot product of $x$ and $y$, and $\| x \|$ and $\| y \|$ are their norms.

### 2. Linkage Criteria

Linkage criteria determine how the distance between clusters is calculated, impacting the cluster shapes and sizes. Common linkage methods include:

- **Single Linkage**: The distance between two clusters $S$ and $T$ is defined as:
  $$
  d(S, T) = \min_{x \in S, y \in T} d(x, y)
  $$

- **Complete Linkage**: The distance between two clusters $S$ and $T$ is defined as:
  $$
  d(S, T) = \max_{x \in S, y \in T} d(x, y)
  $$

- **Average Linkage**: The distance between two clusters $S$ and $T$ is defined as:
  $$
  d(S, T) = \frac{1}{|S| |T|} \sum_{x \in S} \sum_{y \in T} d(x, y)
  $$

- **Ward’s Method**: The distance between two clusters $S$ and $T$ is defined as the increase in variance when they are merged:
  $$
  \Delta \text{Var}(S, T) = \frac{|S| |T|}{|S| + |T|} d(\bar{x}_S, \bar{x}_T)^2
  $$
  where $\bar{x}_S$ and $\bar{x}_T$ are the centroids of clusters $S$ and $T$, respectively, and $d$ is the distance between the centroids.

### 3. Cluster Initialization

Initially, each data point is treated as a separate cluster. If there are $N$ data points, there are $N$ initial clusters:

$$
C_i = \{ x_i \}, \quad i = 1, 2, \dots, N
$$

## 4. Dendrogram

The dendrogram is a visual representation showing the multi-level hierarchy of clusters formed. It records the distances at which clusters merge, typically visualized vertically.

## 5. Termination Criterion

The algorithm stops based on a predefined condition, such as:

- A specific number of clusters is reached.
- The distance between the nearest clusters exceeds a threshold.


## **Algorithmic Steps in Agglomerative Clustering**

### 1. **Initialization**
- Each data point is treated as its own cluster. If there are **N** data points, we start with **N** clusters.

### 2. **Compute Proximity**
- Calculate the distance between every pair of clusters using a distance metric. Common metrics include:
  - **Euclidean distance**: Suitable for geometric shapes.
  - **Manhattan distance**: Used for grid-like distances.
  - **Cosine similarity**: Commonly used for text data or high-dimensional datasets.

### 3. **Merge Clusters**
- Identify the two closest clusters and merge them into a single cluster.
- This reduces the total number of clusters by one.

### 4. **Update Distance Matrix**
- After merging, update the distance matrix to reflect the distance between the new cluster and all remaining clusters.
- Linkage methods determine how the distance between clusters is calculated:
  - **Single Linkage**: Defined as the shortest distance between any member of one cluster and any member of the other.
  - **Complete Linkage**: Defined as the longest distance between any member of one cluster and any member of the other.
  - **Average Linkage**: The average distance between all pairs of members from both clusters.
  - **Ward’s Method**: The increase in the total within-cluster variance when two clusters merge.

### 5. **Repeat**
- Repeat the merging process until all data points form a single cluster or until the desired number of clusters is reached.

### 6. **Assess the clustering using silhoutte scores**
- Silhouette Score: This is a method to assess how well the clusters are formed. It measures how similar each point is to its own cluster compared to other clusters. The silhouette score ranges from -1 (poor clustering) to +1 (good clustering).
  
Note: It’s important to calculate the silhouette score after completing clustering to evaluate the quality of the clustering result.

### 7. **Dendrogram Construction**
- As clusters are merged, a **dendrogram** is created. This diagram shows how data points are grouped into clusters.
- By cutting the dendrogram at a specific distance, you can decide on the number of clusters that best represent the data.

## **Pros of Agglomerative Clustering**

### 1. **Intuitive and Easy to Understand**
- The algorithm is simple and intuitive, progressively merging the closest pairs of clusters.
- The **dendrogram** helps visualize the process, making it easy to interpret and understand.

### 2. **Flexible Linkage Methods**
- The algorithm allows you to choose different **linkage methods** (e.g., **single**, **complete**, **average**, or **Ward’s**) to fit the specific structure of your data.

### 3. **No Need to Specify Cluster Number Initially**
- Unlike **k-means**, **agglomerative clustering** does not require you to specify the number of clusters beforehand.
- The number of clusters can be determined by analyzing the dendrogram and cutting it at the desired level.

### 4. **Can Handle Non-Elliptical Shapes**
- Agglomerative clustering is capable of identifying clusters with **non-elliptical shapes**, unlike centroid-based methods like **k-means** that assume spherical clusters.
- This makes it ideal for complex datasets with irregular cluster shapes.

### 5. **Robust to Noise and Outliers**
- By choosing appropriate linkage methods (like **complete** or **average linkage**), the algorithm can be more robust to **noise** and **outliers**, avoiding their negative impact on clustering.

## **Cons of Agglomerative Clustering**

### 1. **Computational Complexity**
- Agglomerative clustering is computationally expensive, particularly for large datasets.
- The **time complexity** is **O(N³)**, and the **space complexity** is **O(N²)**, where **N** is the number of data points.
- This makes it less suitable for large-scale datasets.

### 2. **Irreversible Steps**
- Once a **merging decision** is made, it cannot be undone without restarting the algorithm.
- Poor early merge decisions can lead to **sub-optimal clustering** since there is no way to backtrack or correct mistakes.

### 3. **Sensitive to Linkage Criteria**
- The choice of **linkage method** can significantly affect the results of clustering.
- **Single linkage**, in particular, can result in a **chaining effect**, where clusters are linked together inappropriately, leading to less meaningful clusters.

### 4. **Difficulty in Handling Different Sized Clusters and Variances**
- Agglomerative clustering struggles when clusters have very different **sizes**, **numbers of samples**, or **variances**.
- Methods like **single linkage** may perform poorly in such cases, causing the algorithm to miss important cluster structure.

### 5. **No Global Objective**
- Unlike **k-means**, which optimizes a global objective function, agglomerative clustering makes **local decisions**.
- This can sometimes lead to **less cohesive clusters** and prevent the discovery of a **globally optimal structure**.

## **Applications of Agglomerative Clustering**

Agglomerative clustering is widely used across various domains, particularly where hierarchical relationships in the data are important. Key applications include:

1. **Customer Segmentation**: Grouping customers based on purchasing behavior, demographics, or other features.
2. **Image Segmentation**: Dividing an image into segments that represent different regions or objects.
3. **Document Clustering**: Grouping similar documents or articles based on content or keywords.
4. **Anomaly Detection**: Identifying outliers or anomalies in the data, such as fraud detection or network security.
5. **Gene Expression Analysis**: Grouping genes with similar expression patterns in biological research.

# Basic Example of Agglomerative Clustering

## 1. Load the necessary packages and create data

We will be using the make blobs function present in the sklearn library to create synthetic data to demonstrate our basic example of agglomerative clustering.

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score, make_scorer
from sklearn.datasets import make_blobs
from sklearn.model_selection import GridSearchCV
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import pdist

sns.set(style="whitegrid")

X, _ = make_blobs(n_samples=300, centers=4, cluster_std=1.8, random_state=42)  # Increased std for overlap
````

## 2. Plotting the data

````python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], edgecolors='k')
plt.title("Raw Data Before Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
````

## 3. Fitting the model

When the agglomerative clustering model is fit without defining any parameters, it fits the data into 2 clusters by default and uses ward linkage method. This is not always the optimal result that we want.

````python
model = AgglomerativeClustering()
pred = model.fit_predict(X)

score = silhouette_score(X, pred)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=pred, cmap='viridis', edgecolors='k')
plt.title('Agglomerative Clustering')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.colorbar(label="Cluster Label")
plt.show()

print('')
print('Silhouette Score:', score)
````

## 3. Performing hyperparamter tuning

Unlike supervised learning models, Agglomerative Clustering is an unsupervised algorithm, meaning it does not have predefined labels (y_true) for evaluation. GridSearchCV is primarily designed for supervised learning, where it splits data into training and validation sets and evaluates model performance using metrics like accuracy or RMSE. However, since Agglomerative Clustering does not have a predict() method, GridSearchCV cannot directly evaluate clustering quality. Instead, we must assess clustering using an unsupervised metric, such as the Silhouette Score, which measures how well-separated the clusters are.

To tune hyperparameters (n_clusters and linkage method), we implemented manual hyperparameter tuning by iterating through a predefined grid of parameter values. For each combination of n_clusters (3 to 6) and linkage methods (ward, complete, average), we fitted an AgglomerativeClustering model, assigned cluster labels, and computed the Silhouette Score. The best-performing model was selected based on the highest Silhouette Score, ensuring optimal separation between clusters. This manual tuning approach effectively replaces GridSearchCV, allowing us to optimize clustering performance in an unsupervised setting.

````python
n_clusters_options = [3, 4, 5, 6]
linkage_options = ['ward', 'complete', 'average']

best_score = -1
best_params = None
best_labels = None

for n_clusters in n_clusters_options:
    for linkage_method in linkage_options:
        # Fit Agglomerative Clustering
        clustering = AgglomerativeClustering(n_clusters=n_clusters, linkage=linkage_method)
        labels = clustering.fit_predict(X)

        # Compute silhouette score
        score = silhouette_score(X, labels)
        print(f"Clusters: {n_clusters}, Linkage: {linkage_method}, Silhouette Score: {score:.4f}")

        # Save best model
        if score > best_score:
            best_score = score
            best_params = {'n_clusters': n_clusters, 'linkage': linkage_method}
            best_labels = labels
````

````python
print(f"\nBest Parameters: {best_params}")
print(f"Best Silhouette Score: {best_score:.4f}")
````

## 4. Plotting the clusters

````python
def plot_clusters(X, labels, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolors='k')
    plt.title(title)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.colorbar(label="Cluster Label")
    plt.show()

plot_clusters(X, best_labels, title=f"Agglomerative Clustering (Best: {best_params})")
````

## 5. Plotting the Dendogram

````python
def plot_dendrogram(X, linkage_method):
    plt.figure(figsize=(10, 10))

    if linkage_method == 'ward':
        linked = linkage(X, method='ward')
    else:
        linked = linkage(pdist(X), method=linkage_method)

    dendrogram(linked, truncate_mode='level', p=10)
    plt.title(f"Dendrogram ({linkage_method.capitalize()} Linkage)")
    plt.xlabel("Data Points")
    plt.ylabel("Distance")
    plt.show()

plot_dendrogram(X, linkage_method=best_params['linkage'])
````

# Real World Example: Chicago Airbnb Listings

## Loading the dataset

````python
import pandas as pd
df = pd.read_csv('listings.csv')
````

## Exploratory Data Analysis

````python
df.head()
````

````python
df.info()
df.isnull().sum()
````

````python
df_subset = df[['latitude', 'longitude', 'price', 'room_type', 'review_scores_rating', 'number_of_reviews']]
df_subset.head()
````

````python
df_subset.loc[:, 'price'] = df_subset['price'].replace('[\$,]', '', regex=True).astype(float)
df_subset.head()
````

````python
df_subset.isnull().sum()
````

````python
df_subset = df_subset.dropna(subset=['price'])

median_rating = df_subset['review_scores_rating'].median()
df_subset.loc[:,'review_scores_rating'] = df_subset['review_scores_rating'].fillna(median_rating)

df_subset.isnull().sum()
````

### Encoding Room Type (Categorical Feature)

````python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_subset['room_type_encoded'] = le.fit_transform(df_subset['room_type'])
df_subset = df_subset.drop(columns=['room_type'])

df_subset.head()
````

````python
room_type_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(room_type_mapping)
````

### Normalizing the variables

````python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

columns_to_scale = ['latitude', 'longitude', 'price', 'review_scores_rating', 'number_of_reviews', 'room_type_encoded']
df_scaled = scaler.fit_transform(df_subset[columns_to_scale])
df_scaled = pd.DataFrame(df_scaled, columns=columns_to_scale)

df_scaled.head()
````

## Agglomerative Clustering

### Initially Starting with 4 variables

````python
from sklearn.cluster import AgglomerativeClustering

agg_clustering = AgglomerativeClustering(n_clusters=4, linkage='ward')
df_scaled['cluster'] = agg_clustering.fit_predict(df_scaled)
df_scaled.head()
````

### Visualizing the clusters

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(df_scaled['longitude'], df_scaled['latitude'], c=df_scaled['cluster'], cmap='viridis', alpha=0.5)

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Agglomerative Clustering of Airbnb Listings (Geospatial View)")
plt.colorbar(label="Cluster")

plt.show()
````
